When performing data analysis, you should be very aware of which columns have missing values.  Doing so will help you be careful about using COUNT(\*) or COUNT(col_name) in appropriate places.

Generate a table showing how many nulls there are per column.  

Rename the columns as shown.

So your final answer should match the table below, exactly.

| missing_donation | missing_state | missing_affiliation | missing_class_year | missing_preferred_division | missing_date_submitted | missing_donation_year |

# QUESTION 3

Determine how many contributors there were per state.

Order your result from the state with the highest number of contributors to the lowest.  

Rename the column with the number of contributors to appear as ``num_contributors``.



````python
# CodeGrade Tag: question_3
sql_statement3 = """ SELECT state, COUNT(*) AS num_contributors
                     From hearts_on_fire
                     GROUP BY state
                     Order BY num_contributors DESC;"""

query_results3 = pd.read_sql_query(sql_statement3, conn)
query_results3
````

# QUESTION 5

Recall that the Northeast Region is CT, ME, MA, NH, RI, VT, NJ, NY, PA.
  
Write a query to determine which states OUTSIDE of the Northeast Region had the 3 largest amounts of alumni contributors in the year 2021.

Your columns should be pertinent to the question at hand and your final table should only have 3 rows, making your answer clear.


````python
# CodeGrade Tag: question_5
sql_statement5 = """SELECT state, affiliation, COUNT(*) AS num_contributors, donation_year
                    FROM hearts_on_fire
                    WHERE state NOT IN ('CT', 'ME', 'MA', 'NH', 'RI', 'VT', 'NJ', 'NY', 'PA') 
                      AND donation_year = 2021
                      AND affiliation = 'Alumni'
                    GROUP BY state
                    ORDER BY num_contributors DESC
                    LIMIT 3;"""

query_results5 = pd.read_sql_query(sql_statement5, conn)
query_results5
````

# QUESTION 6

For the year 2020, show the number of donors per the amount of donation.  

In the final results, order the donations from smallest to largest.

````python
# CodeGrade Tag: question_6
sql_statement6 = """SELECT COUNT(affiliation) AS num_donors, donation AS donation_amount, donation_year
                    FROM hearts_on_fire
                    WHERE donation_year = 2020
                    GROUP BY donation_amount
                    ORDER BY donation_amount ASC;"""

query_results6 = pd.read_sql_query(sql_statement6, conn)
query_results6
````

# QUESTION 7

There were a lot of 1s in the prior table.  

Run the same query but this time, show only those rows where more than 1 person contributed the given amount.

````python
# CodeGrade Tag: question_7
sql_statement7 = """SELECT COUNT(affiliation) AS num_donors, donation AS donation_amount, donation_year
                    FROM hearts_on_fire
                    WHERE donation_year = 2020
                    GROUP BY donation_amount
                    HAVING num_donors > 1
                    ORDER BY donation_amount ASC;"""


query_results7 = pd.read_sql_query(sql_statement7, conn)
query_results7
````

# QUESTION 8

# QUESTION 9

Show the average donations per alumni class year, ordered from highest average to lowest. 

Name the average column:  ```avg_donation```.

````python
# CodeGrade Tag: question_9
sql_statement9 = """SELECT AVG(donation) AS avg_donation, affiliation, class_year
                    FROM hearts_on_fire
                    WHERE affiliation = 'Alumni'
                    GROUP BY class_year
                    ORDER BY avg_donation DESC;"""

query_results9 = pd.read_sql_query(sql_statement9, conn)
query_results9
````
